```python
import cv2
import os

# Set your working directories
IMAGE_PATH =  # Your large image to patch 
DATA_DIR = # Your directory to save patches

PATCH_SIZE = 256

#Set which training image number this is i.e. 1, 2, 3, 4
#this creates your patch file names for the iteration you are on i.e. patch_1_1...patch_1_30
image_number = 1
```

```python
def create_patches(image_path, patch_size, output_dir):
    img = cv2.imread(image_path, 0) # Load as grayscale
    h, w = img.shape
    
    # Create folders
    os.makedirs(f"{output_dir}/patches_{image_number}", exist_ok=True)
    
    patches = []
    patch_count = 0
    
    # Patch image
    for y in range(0, h - patch_size + 1, patch_size):
        for x in range(0, w - patch_size + 1, patch_size):
            patch = img[y:y+patch_size, x:x+patch_size]
            
            # Save patch
            patch_name = f"patch_{image_number}_{patch_count}.jpg"
            cv2.imwrite(f"{output_dir}/patches_{image_number}/{patch_name}", patch)
            patch_count += 1
            
    print(f"Created {patch_count} patches")

create_patches(IMAGE_PATH, PATCH_SIZE, DATA_DIR)
```
