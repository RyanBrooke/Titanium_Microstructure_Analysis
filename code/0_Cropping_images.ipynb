```python
import cv2
import os
import glob

## Crops out the information panel from the SEM micrograph

# ================= CONFIGURATION =================
# 1. Folder containing your original images
INPUT_FOLDER = ## input folder path 

# 2. Folder where you want the cropped versions saved
OUTPUT_FOLDER = ## output folder path

# 3. Crop Settings (from your snippet)
x_start = 0
y_start = 0
height = 1792
width = 2560
# =================================================

# Create the output folder if it doesn't exist
if not os.path.exists(OUTPUT_FOLDER):
    os.makedirs(OUTPUT_FOLDER)
    print(f"Created folder: {OUTPUT_FOLDER}")

# Find all images (jpg, png, tif)
# You can add more extensions to this list if needed
extensions = ["*.jpg", "*.png", "*.tif", "*.tiff", "*.bmp"]
image_files = []
for ext in extensions:
    image_files.extend(glob.glob(os.path.join(INPUT_FOLDER, ext)))

print(f"Found {len(image_files)} images. Starting crop...")

for filepath in image_files:
    # Read the image
    img = cv2.imread(filepath)
    
    if img is None:
        print(f"Skipping {filepath} (could not be read)")
        continue

    # Get the dimensions of the loaded image to prevent crashing 
    # if an image is smaller than the crop size
    img_h, img_w = img.shape[:2]
    
    # Optional: Safety check
    if img_h < (y_start + height) or img_w < (x_start + width):
        print(f"Warning: {os.path.basename(filepath)} is smaller than the crop size. Cropping to max available.")
    
    # Perform the crop
    # Slicing: [rows (y), cols (x)]
    crop = img[y_start : y_start+height, x_start : x_start+width]
    
    # Construct output path
    filename = os.path.basename(filepath) # e.g., "sample_01.jpg"
    save_path = os.path.join(OUTPUT_FOLDER, filename)
    
    # Save
    cv2.imwrite(save_path, crop)
    print(f"Saved: {filename}")

print("Batch processing complete.")
```
